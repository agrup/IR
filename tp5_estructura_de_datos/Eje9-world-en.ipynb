```python
import matplotlib.pyplot as plt

import numpy as np

file_words = '../colecciones/words-en.txt'

total = 0
#total
sizes_words = []
max_size = 0
with open(file_words) as file:
    lines = file.readlines()
    for line in lines:
        len_w = len(line)
        if len_w > max_size:
            max_size = len_w
        
        total += len_w
        sizes_words.append(len_w)
avg = total/ len(sizes_words)

%matplotlib inline


plt.xlabel('Largo')
plt.ylabel('Cantidad de palabras')
#plt.hist(sizes_words, bins=max_size, color="red")
y,hist = np.histogram(sizes_words,bins=10)
centros=(hist[1:]+hist[:-1])
plt.plot(centros,y,'-',color='red')

plt.show()
print("Cantidad de palabras: ", len(sizes_words))

print("Palabra mas larga: ", max_size)
print("Media: ", avg)
print("Total size: ", total)
```
